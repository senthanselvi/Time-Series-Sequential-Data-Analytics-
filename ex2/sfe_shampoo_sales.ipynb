```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split
```

```python
# Step 1: Download the Shampoo Sales dataset
sd = pd.read_csv('shampoo_sales.csv')
sd
```

```python
# Step 2: EDA
plt.figure(figsize = (10,6))
plt.plot(sd['Month'],sd['Sales'],marker = 'o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Shampoo Sales Trend')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
# Calculate summary statistics
summary_stats = sd['Sales'].describe()
print("Summary Statistics:\n",summary_stats)
```

```python
sd.info()
```

```python
# Outlier detection (example using IQR method)
Q1 = sd['Sales'].quantile(0.25)
Q3 = sd['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
sd_out = sd[(sd['Sales'] <= lower_bound) & (sd['Sales'] >= upper_bound)]
sd_out
```

```python
# Correlation analysis
numeric_sd = sd.select_dtypes(include=['int64', 'float64'])
correlation = numeric_sd.corr()
print("Correlation Matrix:\n", correlation)
 
```

```python
# Calculate rolling mean with a window size of 3 months
sd['rolling_mean'] = sd['Sales'].rolling(window = 3).mean()

plt.figure(figsize=(10,6))
plt.plot(sd['Month'],sd['Sales'],label = 'Sales')
plt.plot(sd['Month'],sd['rolling_mean'],label='Rolling Mean(3 months)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Shampoo Sales with Rolling Mean')
plt.xticks(rotation = 45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.boxplot(sd['Sales'])
plt.title('box Plot of Shampoo Sales')
plt.ylabel('Sales')
plt.show()
```

```python
# Create lagged features
for lag in [1,2,3,6,12]:
    sd[f'lag_{lag}'] = sd['Sales'].shift(lag)

# Drop rows with missing values due to lag creation
sd.dropna(inplace=True)

plt.figure(figsize=(12,6))
for lag in [1,2,3,6,12]:
    plt.plot(sd['Month'],sd[f'lag_{lag}'],label=f'Lag{lag}')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Lagged Sales Features')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 3: Date and Lag-based Features
# Convert 'Month' column to datetime format
sd['Month'] = pd.to_datetime(sd['Month'],format='%m-%d')
sd
```

```python
# Create lag features
sd['sales_lag1'] = sd['Sales'].shift(1)
sd['sales_lag2'] = sd['Sales'].shift(2)
sd
```

```python
# Calculate difference between consecutive sales values
sd['sales_diff1'] = sd['Sales'] - sd['sales_lag1']
sd['sales_diff2'] = sd['Sales'] - sd['sales_lag2']
sd
```

```python
# Extract features from datetime column
sd['year'] = sd['Month'].dt.year
sd['month'] = sd['Month'].dt.month
sd['day_of_week'] = sd['Month'].dt.dayofweek
sd
```

```python
# Step 4: Sliding and Expanding Window Features
# Create a sliding window feature
window_size = 3  # Number of months in the sliding window
sd['sales_sliding_window'] = sd['Sales'].rolling(window=window_size).mean()
sd
```

```python
# Generate an expanding window feature
sd['sales_expanding_window'] = sd['Sales'].expanding().mean()
sd
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# Load the dataset
data = pd.read_csv("bill charge.csv")

# Feature Engineering
# Extract year, month, and day from the "Date" column
data["Year"] = pd.to_datetime(data["Date"]).dt.year
data["Month"] = pd.to_datetime(data["Date"]).dt.month
data["Day"] = pd.to_datetime(data["Date"]).dt.day

# Calculate the average bill charge for each patient
data["Avg_Bill_Charge"] = data.groupby("Patient Name")["Bill Charge"].transform("mean")

# Create a binary feature for returning customers
data["Returning_Customer"] = data.groupby("Patient Name")["Patient Name"].transform("count") > 1

# Encode categorical variables
categorical_features = ["Patient Name"]
one_hot_encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)
encoded_data = one_hot_encoder.fit_transform(data[categorical_features])
encoded_columns = one_hot_encoder.get_feature_names_out(categorical_features)
data_encoded = pd.DataFrame(encoded_data, columns=encoded_columns)

# Merge encoded data with the original dataset
data = pd.concat([data, data_encoded], axis=1)

# Drop original categorical columns
data.drop(columns=categorical_features, inplace=True)

# Feature Scaling
numerical_features = ["Age", "Bill Charge", "Avg_Bill_Charge"]
scaler = StandardScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])

# Final Feature Set
features_to_use = [
    "Year",
    "Month",
    "Day",
    "Age",
    "Bill Charge",
    "Avg_Bill_Charge",
    "Returning_Customer",
] + list(encoded_columns)

# Select the final feature set
final_data = data[features_to_use]

# Save the processed data
final_data.to_csv("processed_bill_charge_data.csv", index=False)
```
